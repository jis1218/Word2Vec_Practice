```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
```

```python
train = pd.read_csv('./word2vec-nlp-tutorial/labeledTrainData.tsv', header=0, delimiter='\t', quoting=3)
test = pd.read_csv('./word2vec-nlp-tutorial/testData.tsv', header=0, delimiter='\t', quoting=3)
unlabeled_train = pd.read_csv('./word2vec-nlp-tutorial/unlabeledTrainData.tsv', header=0, delimiter='\t', quoting=3)

print ("Read %d labeled train reviews, %d labeled test reviews, " \
 "and %d unlabeled reviews\n" % (train["review"].size,  
 test["review"].size, unlabeled_train["review"].size ))

lemmatizer = WordNetLemmatizer()
```

```python
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords

def review_to_wordlist( review, remove_stopwords=False ):
    # Function to convert a document to a sequence of words,
    # optionally removing stop words.  Returns a list of words.
    #
    # 1. Remove HTML
    review_text = BeautifulSoup(review).get_text()
    #  
    # 2. Remove non-letters
    review_text = re.sub("[^a-zA-Z]"," ", review_text)
    #
    # 3. Convert words to lower case and split them
    words = review_text.lower().split()
    #
    # 4. Optionally remove stop words (false by default)
    if remove_stopwords:
        stops = set(stopwords.words("english"))
        words = [w for w in words if not w in stops]
    #
    # 5. Lemmatizer
    #words = [lemmatizer.lemmatize(token, "v") for token in words]
    
    # 5. Return a list of words
    return(words)
```

```python
# Download the punkt tokenizer for sentence splitting
import nltk.data
nltk.download()   

# Load the punkt tokenizer
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

# Define a function to split a review into parsed sentences
def review_to_sentences( review, tokenizer, remove_stopwords=False ):
    # Function to split a review into parsed sentences. Returns a 
    # list of sentences, where each sentence is a list of words
    #
    # 1. Use the NLTK tokenizer to split the paragraph into sentences
    raw_sentences = tokenizer.tokenize(review.strip())
    #
    # 2. Loop over each sentence
    sentences = []
    for raw_sentence in raw_sentences:
        # If a sentence is empty, skip it
        if len(raw_sentence) > 0:
            # Otherwise, call review_to_wordlist to get a list of words
            sentences.append( review_to_wordlist( raw_sentence, \
              remove_stopwords ))
    #
    # Return the list of sentences (each sentence is a list of words,
    # so this returns a list of lists
    return sentences
```

```python
sentences = []  # Initialize an empty list of sentences

print ("Parsing sentences from training set")
for review in train["review"]:
    sentences += review_to_sentences(review, tokenizer)

print ("Parsing sentences from unlabeled set")
for review in unlabeled_train["review"]:
    sentences += review_to_sentences(review, tokenizer)
```

```python
word_set = set([])
word_dict = dict()
i = 0
for sentence in sentences:
    for word in sentence:
        if not word in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] = word_dict[word] + 1
    
    i+=1
    if i==1000:
        print(i)
        print(len(word_dict))
```

```python
important_words = []
for word in word_dict.keys():
    if word_dict[word]>40:
        important_words.append(word)
```

```python
import numpy as np
```

```python
vector_num = len(important_words)+1
print(vector_num)
hidden_state = 300
first_weight = np.random.rand(vector_num, hidden_state)
print(first_weight)
second_weight = np.random.rand(hidden_state, vector_num)
```

```python
index_dict = dict()
index_dict['NONWORD'] = 0
for i, word in enumerate(important_words):
    index_dict[word] = i+1

print(index_dict)
```

```python
eye = 0
eye = np.eye(vector_num)
print(eye)
print(np.shape(eye))
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
# Skip-Gram
window_size = 5
for sentence in sentences:
    input = ''
    print(sentence)
    for i in range(len(sentence)):
        if not sentence[i] in index_dict:
            continue
        else:
            input = eye[index_dict[sentence[i]]]
        hidden_layer = np.matmul(input, first_weight)
        output = np.matmul(hidden_layer, second_weight)
        softmax_output = softmax(output)
        #print(softmax(output))
        target = np.zeros(vector_num)
        for j in range(int(-window_size/2), int(window_size/2)+1):
            if i+j>-1 and i+j<len(sentence) and j!=0:                
                if not sentence[i+j] in index_dict:
                    target[0] = 1
                else:
                    target[index_dict[sentence[i+j]]]=1
        #print(target)
        print(np.shape(softmax_output))
        print(np.shape(target))
        
        
        
        
    break
```
